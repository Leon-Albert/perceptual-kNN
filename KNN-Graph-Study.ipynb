```python
import networkx as nx

PathGraph = 'data/Knn-G/knnG_test.gml'
G = nx.read_gml(PathGraph)

# ea = nx.get_edge_attributes(G, 'edge_attr')
# for e in ea:
#     if ea[e]>10e6:
#         ea[e] = 10e6

# nx.set_edge_attributes(G,ea,name='weight')

nx.remove_edge_attributes(G,'edge_attr')

G
```

## Visualisation : Bokeh

```python
from bokeh.plotting import figure, from_networkx, show
from bokeh.models import (MultiLine, NodesAndAdjacentNodes, Scatter, TapTool)
from bokeh.palettes import Spectral4
from bokeh.io import output_notebook

output_notebook()

plot = figure(width=800, height=600,
              x_axis_location=None, y_axis_location=None, toolbar_location=None,
              title="K-nn Graph", background_fill_color="#ffffff")
              #,output_backend="webgl")
              #,tooltips="index: @index, x: @x")

plot.grid.grid_line_color = None

plot.add_tools(TapTool(mode='replace'))

graph_renderer = from_networkx(G, nx.spring_layout)

scatter_glyph = Scatter(size=15, fill_color=Spectral4[0])
graph_renderer.node_renderer.glyph = scatter_glyph

graph_renderer.node_renderer.selection_glyph = scatter_glyph.clone(fill_color=Spectral4[1])
graph_renderer.node_renderer.nonselection_glyph = scatter_glyph.clone(fill_color=Spectral4[0])


ml_glyph = MultiLine(line_color="#CCCCCC", line_alpha=0.5, line_width=0.1)
graph_renderer.edge_renderer.glyph = ml_glyph


graph_renderer.selection_policy = NodesAndAdjacentNodes()


plot.renderers.append(graph_renderer)

show(plot)
```
